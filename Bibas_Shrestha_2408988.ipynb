```python
# Filter the dataset for fares greater than 100
fare_gt_100 = df[df['Fare'] > 100]

# Print the result
fare_gt_100.head()
```

```python
# Filter the dataset for females under 18
female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]

# Print the result
female_under_18.head()
```

```python
# Filter the dataset for passengers who embarked at Cherbourg or Southampton
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]

# Print the result
embarked_c_or_s.head()
```

```python
# Filter the dataset for passengers in first or second class
first_second_class = df[df['Pclass'].isin([1, 2])]

# Print the result
first_second_class.head()
```

```python
# Add a new column 'fare_per_year' to the DataFrame
# To avoid division by zero, we can replace 'Age' values of 0 with NaN
df['fare_per_year'] = df['Fare'] / df['Age'].replace(0, pd.NA)

# Print the updated DataFrame to see the new column
df[['Fare', 'Age', 'fare_per_year']].head()
```

```python
# Add a new column 'fare_per_year' to the DataFrame
df['fare_per_year'] = df['Fare'] / df['Age'].replace(0, pd.NA)

# Subset rows where fare_per_year is greater than 5
high_fare_age = df[df['fare_per_year'] > 5]

# Print the result
high_fare_age.head()
```

```python


# Sort high_fare_age by descending fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# Print the sorted result
high_fare_age_srt.head()
```

```python
# Filter the dataset for passengers in first class
first_class = df[df['Pclass'] == 1]

# Print the result
first_class.head()
```

```python
# Select only the Name and fare_per_year columns
result = high_fare_age_srt[['Name', 'fare_per_year']]

# Look at the result
result.head()
```

```python
# Calculate the total Fare paid across all passengers
total_fare = df['Fare'].sum()

# Print the total fare
total_fare
```

```python
# Subset for passengers in first class (Pclass is 1)
first_class_passengers = df[df['Pclass'] == 1]

# Calculate the total fare for first class passengers
total_fare_first_class = first_class_passengers['Fare'].sum()

# Print the total fare
total_fare_first_class
```

```python
# Calculate total fare for second class passengers (Pclass is 2)
second_class_passengers = df[df['Pclass'] == 2]
total_fare_second_class = second_class_passengers['Fare'].sum()
total_fare_second_class

# Calculate total fare for third class passengers (Pclass is 3)
third_class_passengers = df[df['Pclass'] == 3]
total_fare_third_class = third_class_passengers['Fare'].sum()
total_fare_third_class
```

```python
# Combine the fare totals into a list
fare_totals = [total_fare_first_class, total_fare_second_class, total_fare_third_class]

# Print the list of fare totals
fare_totals
```

```python
# Calculate the proportion of fare revenue by class
fare_proportion_by_class = [total_fare_first_class / total_fare, total_fare_second_class / total_fare, total_fare_third_class / total_fare]

fare_proportion_by_class
```

```python
# Create the 'Age Group' column using .loc
df['Age Group'] = 'unknown'  # Default value in case of missing or undefined ages

# Assign 'child' to ages less than 18
df.loc[df['Age'] < 18, 'Age Group'] = 'child'

# Assign 'adult' to ages between 18 and 64
df.loc[(df['Age'] >= 18) & (df['Age'] < 65), 'Age Group'] = 'adult'

# Assign 'senior' to ages 65 and above
df.loc[df['Age'] >= 65, 'Age Group'] = 'senior'

# Check the result
df.head()
```

```python
#Calculate the total number of passengers
total_passengers = df.shape[0]

total_passengers
```

```python
# Count the number of passengers in each age group
age_group_counts = df['Age Group'].value_counts()

age_group_counts
```

```python
#Calculate the proportion of passengers in each age group
age_group_proportions = age_group_counts / total_passengers

age_group_proportions
```

```python
#Convert proportions to percentages
age_group_percentages = age_group_proportions * 100

age_group_percentages
```
